# Experiment 2 â€“ Hidden units as learned features
Visualize each hidden unit's weight vector as a small image.

```python

import numpy as np
import matplotlib.pyplot as plt
from rbm_utils import load_model

params, target_size = load_model("models/rbm_digits_50.npz")
W = params.W
print("Loaded model with weight matrix shape:", W.shape)
```

```python

num_filters = 25
rows = cols = 5
fig, axes = plt.subplots(rows, cols, figsize=(8, 8))
for j in range(num_filters):
    r = j // cols
    c = j % cols
    filt = W[:, j].reshape(target_size)
    fmin, fmax = filt.min(), filt.max()
    filt_norm = (filt - fmin) / (fmax - fmin + 1e-8)
    axes[r, c].imshow(filt_norm, cmap="gray")
    axes[r, c].axis("off")
axes[0, 0].set_title("Filters")
plt.suptitle("Hidden unit filters (columns of W)")
plt.show()
```
